### Functions

```python
#Function seen during the bootcamp,to review
def variable_rate_vs(df, variable, grouper):
    """
    Returns a panda Series with:
    Index: the categories inside the grouper
    Values: the variable (for example Response) rate for the group

    Parameters
    ----------
    df : Pandas Dataframe
        DESCRIPTION.
    variable : String
        The label of the df column containing the variable data used to calculate the rate, e.g Response
    grouper : String
        The label of the df column used to group the variable rate

    Returns
    -------
    Panda Series
        Index: the categories inside the grouper
        Values: the variable (for example Response) rate for the group

    """

    # Dictionary to store the results
    rate_dict = {}
    
    # Series and group seting, they will be used later in the for loop for indexing and accessing the data
    var_cats = df[variable].value_counts()
    group_cats = df[grouper].value_counts()
    group = df.groupby([grouper, variable])[variable].count()
    
    # The outer for loops through the groups
    ## The inner for loops the group series and for the specific group access the variable catergories and stores it in a list
    ## At the end of the innerr loop the rate for the group is stored in a dict of the form {group:rate}
    
    for g in group_cats.index:

        store = []

        for c in var_cats.index:
        
            n = group.loc[(g, c)]
            store.append(n)

        rate_dict[g] = round((store[1] / store[0])*100, 2)

    # A series is returned
    return pd.Series(rate_dict)
```

#### Hospital overall  rating plot and Patient survey star rating plot

```python
star_rating = df["patient survey star rating"].value_counts(dropna=False)
overall_rating = df["hospital overall rating"].value_counts(dropna=False)

fig1, ax1 = plt.subplots(1,2,figsize = (8,8))

ax1[0].set_title('hospital overall rating', fontweight='bold')
ax1[0].bar(overall_rating.index, overall_rating, color=('blue'))

for x,y in zip(overall_rating.head().index,overall_rating.head()):

    label = len(df[df["hospital overall rating"] == x].value_counts())
    ax1[0].annotate(label,(x,y), textcoords="offset points", xytext=(0,10),ha='center')
    


ax1[1].set_title('patient survey star rating', fontweight='bold')
ax1[1].bar(star_rating.index, star_rating, color=("red"))

for x,y in zip(star_rating.head().index,star_rating.head()):

    label = len(df[df["patient survey star rating"] == x].value_counts())
    ax1[1].annotate(label,(x,y), textcoords="offset points", xytext=(0,10),ha='center')
    
plt.savefig("Data/hospital_overall_rating.png",transparent=False)
```

We can notice that the hospital overall rating is more evenly distributed among the notes than the patient survey star rating, which can be explained by the fact that the hospital overall rating adds more features to the patient survey star rating that impact positively or negatively the ratings.

#### Measure ID Terminology

- Measure ID terminology

- a_p = always .
- u_p = usually .
- sn_p = sometimes or never .
- py = probably yes .
- dy = definitely yes .
- pn = probably no .
- dn = definitely no .
- sa = strongly agree
- sd = strongly disagree .
- 0_6 = 0 to 6 .
- 7_8 = 7 to 8 .
- 9_10 = 9 to 10
- y_p = yes 
- n_p = no .
- a = agree

#### Sentiment analyzer

```python
sentiment = []
for x in range(len(df3["hcahps measure id"])):
    if str(df3["hcahps measure id"][x]).endswith("sn_p") == True:
        sentiment.append("negative")
    elif str(df3["hcahps measure id"][x]).endswith("dn") == True:
        sentiment.append("negative")
    elif str(df3["hcahps measure id"][x]).endswith("pn") == True:
        sentiment.append("negative")
    elif str(df3["hcahps measure id"][x]).endswith("sd") == True:
        sentiment.append("negative")
    elif str(df3["hcahps measure id"][x]).endswith("0_6") == True:
        sentiment.append("negative")
    elif str(df3["hcahps measure id"][x]).endswith("n_p") == True:
        sentiment.append("negative")
    elif str(df3["hcahps measure id"][x]).endswith("a_p") == True:
        sentiment.append("positive")
    elif str(df3["hcahps measure id"][x]).endswith("u_p") == True:
        sentiment.append("positive")
    elif str(df3["hcahps measure id"][x]).endswith("py") == True:
        sentiment.append("positive")
    elif str(df3["hcahps measure id"][x]).endswith("dy") == True:
        sentiment.append("positive")
    elif str(df3["hcahps measure id"][x]).endswith("7_8") == True:
        sentiment.append("positive")
    elif str(df3["hcahps measure id"][x]).endswith("9_10") == True:
        sentiment.append("positive")
    elif str(df3["hcahps measure id"][x]).endswith("y_p") == True:
        sentiment.append("positive")
    elif str(df3["hcahps measure id"][x]).endswith("a") == True:
        sentiment.append("positive")
    else:
        sentiment.append("unknown")
```

```python
df3["sentiment"] = sentiment
```

#### Rating over service in percentage

```python
#positive sentiment

avg_patient_positive = round(df3[df3["sentiment"] == "positive"].groupby(['hcahps answer description'])['hcahps answer percent'].mean(),2)

#plotting figures 

fig1, ax1 = plt.subplots(1,2,figsize = (20,10))

#sorting values

df_sorted = avg_patient_positive.sort_values(ascending=False,inplace =True)

#patient satisfaction

ax1[0].set_title('patient survey satisfaction in percentage, per question', fontweight='bold')
ax1[0].bar(avg_patient_positive.head().index, avg_patient_positive.head(), data = df_sorted,color=('blue'))
ax1[0].tick_params(axis='x', labelrotation = 80)

for x,y in zip(avg_patient_positive.index,avg_patient_positive):

    label = len(df3[df3["hcahps answer description"] == x].value_counts())
    ax1[0].annotate(label,(x,y), textcoords="offset points", xytext=(0,10),ha='center')
    
#negative sentiment
    
avg_patient_negative = round(df3[df3["sentiment"] == "negative"].groupby(['hcahps answer description'])['hcahps answer percent'].mean(),2)

#sorting values 

df_sorted = avg_patient_negative.sort_values(ascending=False,inplace =True)

#patient insatisfaction

ax1[1].set_title('patient survey insatisfaction in percentage, per question', fontweight='bold')
ax1[1].bar(avg_patient_negative.head().index, avg_patient_negative.head(), data = df_sorted,color=('cyan'))
ax1[1].tick_params(axis='x', labelrotation = 80)

for x,y in zip(avg_patient_negative.index,avg_patient_negative):

    label = len(df3[df3["hcahps answer description"] == x].value_counts())
    ax1[1].annotate(label,(x,y), textcoords="offset points", xytext=(0,10),ha='center')
    
plt.savefig("Data/patient_satisfaction_per_question.png",transparent=False)
```

- yes, staff "did" give patients this information = discharge information

- positive : staff communication, courtesy and respect from doctors and nurses

- Staff "always" explained : communication about medications

- negative ; side effects explanation, discharge explanation , quietness
